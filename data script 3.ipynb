```python
import pandas as pd
import pandasql as ps

# Load your feature-engineered CSV
df = pd.read_csv('/content/zomato_featured.csv')
```

```python
query = """
SELECT restaurant_name, area, revenue, profit
FROM df
ORDER BY revenue DESC
LIMIT 10
"""
top_revenue = ps.sqldf(query, locals())
top_revenue
```

```python
query = """
SELECT restaurant_name, area, profit, profit_margin
FROM df
ORDER BY profit DESC
LIMIT 10
"""
top_profit = ps.sqldf(query, locals())
top_profit
```

```python
query = """
SELECT area, SUM(revenue) AS total_revenue, SUM(profit) AS total_profit
FROM df
GROUP BY area
ORDER BY total_revenue DESC
"""
area_revenue = ps.sqldf(query, locals())
area_revenue
```

```python
query = """
SELECT restaurant_type, AVG(rate_out_of_5) AS avg_rating
FROM df
GROUP BY restaurant_type
ORDER BY avg_rating DESC
"""
rating_by_type = ps.sqldf(query, locals())
rating_by_type
```

```python
query = """
SELECT cuisines_type, SUM(revenue) AS total_revenue
FROM df
GROUP BY cuisines_type
ORDER BY total_revenue DESC
LIMIT 10
"""
top_cuisines = ps.sqldf(query, locals())
top_cuisines
```

```python
# Save tables as CSV
top_revenue.to_csv('top_revenue.csv', index=False)
top_profit.to_csv('top_profit.csv', index=False)
area_revenue.to_csv('area_revenue.csv', index=False)
rating_by_type.to_csv('rating_by_type.csv', index=False)
top_cuisines.to_csv('top_cuisines.csv', index=False)

print("All tables saved as CSV")
```
