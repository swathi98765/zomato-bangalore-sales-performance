```python
import pandas as pd

# Load CSV
df = pd.read_csv('/content/zomato_cleaned.csv')

# Business assumptions
"""
1. Number of ratings ≈ number of orders
2. Avg cost (two people) ≈ order value
3. Zomato commission + operational cost = 70% of revenue
"""

# Feature Engineering
df['orders'] = df['rate_out_of_5']
df['revenue'] = df['orders'] * df['avg_cost_two_people']
df['cost'] = df['revenue'] * 0.70
df['profit'] = df['revenue'] - df['cost']
df['profit_margin'] = (df['profit'] / df['revenue']) * 100

# Quick check
print(df[['orders', 'revenue', 'profit', 'profit_margin']].head())

# Save featured CSV
df.to_csv('zomato_featured.csv', index=False)
print("Feature engineering completed. File saved as zomato_featured.csv")
```
